### Step 1: Generate a GitHub Personal Access Token (PAT)

To interact with GitHub from Colab, you need a Personal Access Token (PAT). Follow these steps to generate one:

1.  **Go to GitHub Settings**: Navigate to your GitHub profile settings.
2.  **Developer Settings**: On the left sidebar, click on `Developer settings`.
3.  **Personal access tokens**: Click on `Personal access tokens` and then `Tokens (classic)`.
4.  **Generate New Token**: Click on `Generate new token` and then `Generate new token (classic)`.
5.  **Note**: Give your token a descriptive name (e.g., `Colab_Access`).
6.  **Expiration**: Set an appropriate expiration date (e.g., 90 days or custom).
7.  **Select Scopes**: Crucially, select the `repo` scope (this includes `repo:status`, `repo_deployment`, `public_repo`, `repo:invite`, `security_events`). This gives the token permission to create, read, update, and delete your repositories.
8.  **Generate token**: Click the `Generate token` button.
9.  **Copy the Token**: **IMPORTANT**: Copy the generated token immediately and save it somewhere secure. You will not be able to see it again after you leave the page.

### Step 2: Store your GitHub PAT securely in Colab Secrets

Once you have your token, you should store it securely in Colab's Secrets manager. This prevents your token from being exposed in your notebook code.

1.  Click on the '' icon (Secrets) in the left-hand panel of Colab.
2.  Click `+ New secret`.
3.  Set the Name to `GITHUB_TOKEN`.
4.  Paste your copied GitHub PAT into the Value field.
5.  Ensure `Notebook access` is toggled ON for this secret.

After you've done this, run the cell below to load your token.

```python
from google.colab import userdata

# Retrieve the GitHub token from Colab secrets
GITHUB_TOKEN = userdata.get('GITHUB_TOKEN')

if not GITHUB_TOKEN:
    raise ValueError("GitHub Token not found. Please add it to Colab Secrets as 'GITHUB_TOKEN'.")

print("GitHub Token loaded successfully!")
```

### Step 3: Upload the notebook to GitHub

First, we need to find the name of this notebook file. Then, we'll clone your GitHub repository, copy this notebook into it, and push the changes.

```python
import os
import subprocess

# Find the current notebook's filename
# This assumes the current working directory contains the .ipynb file
notebook_files = [f for f in os.listdir('.') if f.endswith('.ipynb')]
if not notebook_files:
    raise FileNotFoundError("No .ipynb file found in the current directory.")

# Assuming this is the current notebook you want to upload
current_notebook_name = notebook_files[0]
print(f"Detected notebook name: {current_notebook_name}")

# Define your GitHub repository details
REPO_OWNER = 'kapil-bool'
REPO_NAME = 'deep-learning'
REPO_URL = f'https://github.com/{REPO_OWNER}/{REPO_NAME}.git'

# Clone the repository using the GITHUB_TOKEN for authentication
# The token is read from Colab secrets in the previous step
# This will allow us to push changes back to the repository.

# Construct the authenticated URL
authenticated_repo_url = f'https://{GITHUB_TOKEN}@github.com/{REPO_OWNER}/{REPO_NAME}.git'

# Check if the repository directory already exists, and remove it to ensure a clean clone
if os.path.exists(REPO_NAME):
    print(f"Removing existing {REPO_NAME} directory...")
    subprocess.run(['rm', '-rf', REPO_NAME], check=True)

print(f"Cloning {REPO_URL}...")
subprocess.run(['git', 'clone', authenticated_repo_url], check=True)

# Copy the current notebook into the cloned repository
notebook_destination_path = os.path.join(REPO_NAME, current_notebook_name)
print(f"Copying {current_notebook_name} to {notebook_destination_path}...")
subprocess.run(['cp', current_notebook_name, notebook_destination_path], check=True)

# Change directory to the cloned repository
os.chdir(REPO_NAME)

# Configure Git user details for the commit
subprocess.run(['git', 'config', 'user.email', 'colab_user@google.com'], check=True)
subprocess.run(['git', 'config', 'user.name', 'Colab User'], check=True)

# Add the notebook to the staging area
print(f"Adding {current_notebook_name} to git...")
subprocess.run(['git', 'add', current_notebook_name], check=True)

# Commit the changes
commit_message = f"Add {current_notebook_name} from Colab"
print(f"Committing with message: '{commit_message}'...")
subprocess.run(['git', 'commit', '-m', commit_message], check=True)

# Push the changes to GitHub
print("Pushing changes to GitHub...")
subprocess.run(['git', 'push'], check=True)

# Change back to the original directory (optional, but good practice)
os.chdir('..')

print(f"Notebook '{current_notebook_name}' successfully uploaded to {REPO_URL}!")
```
