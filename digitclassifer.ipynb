```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
# data set is already in keras
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train
```

```python
import matplotlib.pyplot as plt
# plt.imshow(x_train[10])
plt.imshow(x_train[0])
# plt.imshow(x_train[15])
```

```python
x_train=x_train/255
x_test=x_test/255
# normalizing the pixel
```

```python
x_train[0]
```

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28))) #a sequention of the 28 next to 28 in a row for input for the first layer
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax')) # multiple output hai so softmax is there
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,epochs=7,validation_split=0.2)
```

```python
y_prob=model.predict(x_test)
```

```python
y_pred=y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
