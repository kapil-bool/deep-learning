```python
def initalize_parameters(layer_dims):
  np.random.seed(3)
  parameters={}
  L=len(layer_dims)
  for l in range(1,L):
    parameters['W'+str(l)]=np.ones((layer_dims[l-1],layer_dims[l]))*0.1
    parameters['b'+str(l)]=np.zeros((layer_dims[l],1)) # Bias should be (neurons, 1)
  return parameters
```

```python
initalize_parameters([2,2,1])
# 2 input 2 neuron in between and the final one giving the output
```

```python
def linear_forward(A_prev,W,b):
  Z=np.dot(W.T,A_prev)+b
  return Z
```

```python
def L_layer_forward(X,parameters):
  A=X; # matrix havind previos output i guess as they will be used here
  L=len(parameters)//2
  for l in range(1,L+1):
    A_prev=A
    Wl=parameters['W'+str(l)]
    bl=parameters['b'+str(l)]
    print("A"+str(l-1)+":",A_prev)
    print("W"+str(l)+":",Wl)
    print("b"+str(l)+":",bl)
    print("**"*34)
    A=linear_forward(A_prev,Wl,bl) #next layer data done

  return A,A_prev
    #both will be used for correction;
```

```python
X=df[['cgpa','resume']].values[0].reshape(2,1)
y=df[['lpa']].values[0][0]
parameters=initalize_parameters([2,2,1])
y_hat,a1=L_layer_forward(X,parameters)
```

```python
yhat=y_hat[0][0]
```

```python
def update_parameters(parameters,a1,y,y_hat,X):
  #total 9
  parameters['W2'][0][0]=parameters['W2'][0][0]+(0.001*2*(y-y_hat)*a1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat)*a1[1][0])
  parameters['b2'][0][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[0][0])
  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0])

  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0])
```

```python
update_parameters(parameters,a1,y,y_hat,X)
```

```python
# //final whole loop implentation
parameters=initalize_parameters([2,2,1])
epochs=5
for i in range(epochs):
  store=[]
  for j in range(df.shape[0]):
    X=df[['cgpa','resume']].values[j].reshape(2,1)
    y=df[['lpa']].values[j][0]
    # now y propgate it forward
    y_hat,A1= L_layer_forward(X,parameters)
    y_hat=y_hat[0][0]
    update_parameters(parameters,A1,y,y_hat,X)
    # store.append(y_hat)
    # here store the mse
    store.append((y_hat-y)**2)

  print('Epoch - ',i+1,'Loss - ',np.array(store).mean())
```
